```python
import pandas as pd

df = pd.read_csv("naver.csv")
clean_df = df.dropna()

genre = clean_df['genre'].value_counts(normalize=True)

most_common_genre = genre.idxmax()
most_common_genre_ratio = genre.max()

print(f"가장 높은 비율을 차지하는 장르: {most_common_genre} ({most_common_genre_ratio:.2%})")
```

```python
import pandas as pd

df = pd.read_csv("naver.csv")
clean_df = df.dropna()

free_counts = clean_df['free'].value_counts(normalize=True)

free_ratio = free_counts.get(True, 0)
paid_ratio = free_counts.get(False, 0)

ratio_difference = abs(free_ratio - paid_ratio)

print(f"무료 비율: {free_ratio:.2%}")
print(f"유료 비율: {paid_ratio:.2%}")
print(f"비율 차이: {ratio_difference:.2%}")
```
