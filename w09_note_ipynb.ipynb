### 01. 데이터 가져오기

```python
#@title CSV 파일 내용 가져오기
with open("weather_data.csv") as data_file:
  data=data_file.readlines()
  print(data)
```

```python
#@title 데이터 정리하기
import csv
with open("weather_data.csv") as data_file:
    data = csv.reader(data_file)
    print(data)
    for row in data:
        print(row)
```

```python
#@title 온도만 별도의 리스트로 저장하기
with open("weather_data.csv") as data_file:
    data = csv.reader(data_file)
    temperatures = []
    print(data)
    for row in data:
        if row[1] != 'temp':
            temperatures.append(row[1])
    print(temperatures)
    temp=0
    for i in range(len(temperatures)):
      temp+=int(temperatures[i])
    temp=round(float(temp/len(temperatures)),2)
    print(f"평균온도 {temp}")

```

```python
#@title pandas 로 csv 파일 읽어오기
import pandas

data = pandas.read_csv("weather_data.csv")
print(data)
```

```python
print(type(data["temp"]))
```

```python
#@title dataframe 을 dictionary 로 저장하기
data_dict=data.to_dict()
print(data_dict)
print(type(data_dict))
```

```python
#@title series 를 list 로 저장하기
temp_list=data["temp"].to_list()
print(temp_list)
print(type(temp_list))
```

```python
#@title 평균 온도 계산하기
print(data["temp"].mean())
```

```python
#@title 온도의 최고점, 최저점, 중간값
print(data["temp"].max())
print(data["temp"].min())
print(data["temp"].median())
```

```python
#@title 금요일의 온도와 상태
print(data[data["day"]=="Friday"])
```

```python
#@title 온도가 가장 높은 요일의 상태
print(data[data["temp"]==data["temp"].max()])
```

```python
#@title 월요일의 온도를 화씨로 변경해서 출력
Monday=data[data["day"]=="Monday"]
Monday_temp=int(Monday.temp.iloc[0])
Monday_temp=(Monday_temp*9/5)+32
print(Monday_temp)
```
